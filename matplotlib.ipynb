### 3. 基本绘图

```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制折线图
plt.plot(x, y, label='Prime Numbers', marker='o')

# 添加标题和标签
plt.title('Line Chart Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 5, 8, 6]

# 绘制柱状图
plt.bar(categories, values, color='skyblue')

# 添加标题和标签
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # 突出显示第一个部分

# 绘制饼图
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# 添加标题
plt.title('Pie Chart Example')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制散点图
plt.scatter(x, y, color='red', marker='x')

# 添加标题和标签
plt.title('Scatter Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# 显示图表
plt.show()
```

### 4. 自定义图表

```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制折线图(修改线条和标记)
plt.plot(x, y, label='Prime Numbers', color='green', linestyle='--', marker='o', linewidth=2, markersize=8)

# 添加标题和标签
plt.title('Customized Line Chart Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制折线图
plt.plot(x, y, label='Prime Numbers', marker='o')

# 添加网格
plt.grid(True)

# 添加注释
plt.annotate('Highest Point', xy=(5, 11), xytext=(4, 10), arrowprops=dict(facecolor='black', shrink=0.05))

# 添加标题和标签
plt.title('Line Chart with Annotations Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]

# 创建一个2行1列的子图
plt.subplot(2, 1, 1)
plt.plot(x, y1, label='Square Numbers', color='blue')
plt.title('First Subplot')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y2, label='Cubic Numbers', color='red')
plt.title('Second Subplot')
plt.legend()

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()
```

### 5. 保存图表

```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制折线图
plt.plot(x, y, label='Prime Numbers', marker='o')

# 添加标题和标签
plt.title('Line Chart Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

# 保存图表为文件
plt.savefig('line_chart.png')

# 显示图表
plt.show()
```
